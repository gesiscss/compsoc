## 2. Data Collections

The [introduction](introduction.ipynb) has laid out why digital behavioral data is catalyzing new social research into human behavior, presented a methodological framework and vocabulary for this endeavor, and stressed the importance of datasets rich in node attributes and types of tie. Here, we dive deeper into the notion of digital bevavioral data. We develop a unified data model that various sets of digital behavioral data can be mapped to. This is a necessary step for methods to be broadly applicable.
